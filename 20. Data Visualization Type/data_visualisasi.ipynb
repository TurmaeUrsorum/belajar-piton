# Types of Data Visualization

**# Types of Data Visualization** berarti "**Jenis-jenis Visualisasi Data**". Ini adalah kategori atau cara berbeda dalam menampilkan data secara visual agar mudah dipahami.

Visualisasi data membantu kita melihat pola, tren, dan hubungan dalam data yang mungkin sulit terlihat hanya dalam bentuk tabel atau teks.

Berikut adalah beberapa **jenis visualisasi data yang umum**:

---

### 1. **Chart (Diagram/Grafik)**

* **Bar Chart (Diagram Batang)**: untuk membandingkan antar kategori.
* **Line Chart (Diagram Garis)**: menunjukkan tren dari waktu ke waktu.
* **Pie Chart (Diagram Lingkaran)**: menunjukkan proporsi/komposisi.
* **Area Chart**: mirip diagram garis, tapi dengan area berwarna di bawah garis.

---

### 2. **Plot**

* **Scatter Plot**: untuk menunjukkan hubungan antara dua variabel.
* **Bubble Chart**: seperti scatter plot, tapi ukuran titik menunjukkan variabel ketiga.

---

### 3. **Diagram Khusus**

* **Histogram**: untuk distribusi data (sering digunakan untuk statistik).
* **Box Plot (Diagram Kotak)**: menunjukkan median, kuartil, dan outlier.
* **Heatmap**: visualisasi data dalam bentuk warna (biasa untuk korelasi atau intensitas).

---

### 4. **Geographical Maps (Peta)**

* **Choropleth Map**: peta yang menunjukkan data berdasarkan warna wilayah (misalnya, kepadatan penduduk).
* **Dot Map**: menunjukkan lokasi titik-titik data.

---

### 5. **Hierarki & Jaringan**

* **Tree Map**: untuk menampilkan struktur hierarki dengan kotak berukuran proporsional.
* **Network Graph**: untuk relasi antar entitas (seperti jejaring sosial).

---

### 6. **Infografis dan Dashboard**

* Gabungan dari berbagai jenis visualisasi untuk menyampaikan informasi secara ringkas dan menarik.

---

# Line Plot

**Line plot** atau **line chart** adalah jenis visualisasi data yang menggunakan **garis lurus yang menghubungkan titik-titik data** untuk menunjukkan **perubahan nilai dari waktu ke waktu** atau berdasarkan urutan tertentu.

---

### ðŸ”¹ **Ciri-ciri Line Plot:**

* Sumbu **X** biasanya menunjukkan **waktu** atau urutan data (hari, bulan, tahun, dsb.).
* Sumbu **Y** menunjukkan **nilai atau ukuran** (seperti suhu, penjualan, jumlah pengguna, dll).
* Titik-titik data dihubungkan oleh **garis** untuk memperlihatkan tren atau pola.

---

### ðŸ“Š **Contoh Sederhana:**

Misalnya kamu ingin menunjukkan jumlah penjualan selama 5 hari:

| Hari   | Penjualan |
| ------ | --------- |
| Senin  | 10        |
| Selasa | 15        |
| Rabu   | 8         |
| Kamis  | 20        |
| Jumat  | 18        |

Line plot-nya akan menampilkan lima titik (satu untuk tiap hari), dan garis yang menghubungkan mereka akan memperlihatkan bagaimana penjualan naik-turun dari hari ke hari.

---

### âœ… **Kapan Pakai Line Plot?**

Gunakan **line plot** saat kamu ingin:

* Menunjukkan **tren dari waktu ke waktu**.
* Membandingkan **beberapa seri data waktu** (misalnya, penjualan 2 produk).
* Melihat **perubahan secara kontinu**.

---

### ðŸ”§ Tools untuk Membuat Line Plot:

* **Excel**
* **Google Sheets**
* **Python** (pakai `matplotlib` atau `seaborn`)
* **Tableau, Power BI**, dan software visualisasi lainnya

---

# Bar Chart 

**Bar chart** (dalam Bahasa Indonesia: **diagram batang**) adalah jenis visualisasi data yang menggunakan **batang-batang (bars)** untuk mewakili nilai-nilai suatu kategori. Panjang atau tinggi batang menunjukkan **besarnya nilai**.

---

### ðŸ”¹ **Ciri-ciri Bar Chart:**

* Biasanya digunakan untuk **membandingkan antar kategori**.
* Setiap **batang mewakili satu kategori**.
* Bisa vertikal (**vertical bar chart**) atau horizontal (**horizontal bar chart**).
* Panjang/tinggi batang sesuai dengan **nilai kuantitatif** (angka).

---

### ðŸ“Š **Contoh Sederhana:**

Misal kita punya data jumlah pelanggan per hari:

| Hari   | Jumlah Pelanggan |
| ------ | ---------------- |
| Senin  | 20               |
| Selasa | 35               |
| Rabu   | 25               |

Bar chart-nya akan terlihat seperti:

* Batang untuk Senin setinggi 20
* Batang untuk Selasa setinggi 35 (lebih tinggi)
* Batang untuk Rabu setinggi 25

---

### âœ… **Kapan Pakai Bar Chart?**

Gunakan **bar chart** saat kamu ingin:

* Membandingkan beberapa kategori.
* Melihat perbedaan besar antar nilai.
* Menampilkan data yang tidak bersifat waktu (bukan tren).

---

### ðŸ§‘â€ðŸ’» **Contoh Kode Python:**

```python
import matplotlib.pyplot as plt

days = ['Senin', 'Selasa', 'Rabu']
customers = [20, 35, 25]

plt.bar(days, customers)
plt.title("Jumlah Pelanggan per Hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Pelanggan")
plt.show()
```

---


````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.bar(x, y)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.show()
````

Kustomisasi yang tersedia untuk Bar Chart :
- color: Untuk wajah bar
- edgecolor: Warna tepi bilah
- linewidth: Lebar tepi bilah
- width: Lebar bilah

````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.bar(x, y, color='yellow', edgecolor='black',
		linewidth=2)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.show()
````

# Histogram

---

## ðŸ“Š **Apa Itu Histogram?**

**Histogram** adalah jenis grafik yang mirip **bar chart**, tetapi digunakan untuk menunjukkan **distribusi frekuensi data numerik** (angka). Artinya, histogram menjawab pertanyaan:

> ðŸ” "Berapa banyak data yang masuk dalam rentang nilai tertentu?"

---

### ðŸ”¹ **Ciri-ciri Histogram:**

* Menampilkan **seberapa sering (frekuensi)** data muncul dalam **kelompok (bins)** tertentu.
* Sumbu **X** berisi **rentang nilai** (misalnya: 0â€“10, 10â€“20, dst).
* Sumbu **Y** berisi **jumlah data** (frekuensi) dalam rentang tersebut.
* Tidak ada **jarak antar batang**, karena data bersifat **kontinu**.

---

### ðŸ“¦ **Contoh Sederhana:**

Misal kamu punya data `tip`:

```
[1, 2, 2.5, 3, 3, 3.5, 4, 4.5, 5, 6]
```

Kalau kamu pakai histogram dengan bin 3:

* Bin 1â€“3 â†’ 4 data
* Bin 3â€“5 â†’ 5 data
* Bin 5â€“7 â†’ 1 data

Maka histogram-nya akan menunjukkan **jumlah data** dalam tiap rentang itu.

---

### ðŸ§‘â€ðŸ’» **Contoh Kode Python:**

```python
import matplotlib.pyplot as plt

tips = [1, 2, 2.5, 3, 3, 3.5, 4, 4.5, 5, 6]

plt.hist(tips, bins=3, edgecolor='black')
plt.title("Distribusi Tip")
plt.xlabel("Nilai Tip")
plt.ylabel("Jumlah Data")
plt.show()
```

---

### ðŸ†š **Perbedaan Histogram vs Bar Chart:**

| Fitur                   | Histogram             | Bar Chart                 |
| ----------------------- | --------------------- | ------------------------- |
| Jenis Data              | **Numerik kontinu**   | **Kategorikal**           |
| Sumbu X                 | Rentang angka (bin)   | Kategori                  |
| Tujuan                  | Lihat distribusi data | Bandingkan antar kategori |
| Ada celah antar batang? | âŒ Tidak               | âœ… Ya                      |

---


````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['total_bill']


# plotting the data
plt.hist(x)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Frequency')

# Adding label on the x-axis
plt.xlabel('Total Bill')

plt.show()
````

Kustomisasi yang tersedia untuk Histogram

- bins: Jumlah tempat sampah dengan lebar yang sama
- color: Untuk mengubah warna wajah
- edgecolor: Warna tepi
- linestyle: Untuk garis tepi
-  alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)

````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['total_bill']

# plotting the data
plt.hist(x, bins=25, color='green', edgecolor='black',
		linestyle='--', alpha=0.5)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Frequency')

# Adding label on the x-axis
plt.xlabel('Total Bill')

plt.show()
````

# Scatterplot

### ðŸ“Œ Apa Itu **Scatter Plot**?

**Scatter plot** (atau diagram pencar) adalah grafik yang digunakan untuk **menampilkan hubungan antara dua variabel numerik**.
Setiap titik dalam scatter plot mewakili **satu data point**, dengan:

* **Sumbu X**: variabel pertama
* **Sumbu Y**: variabel kedua

---

### ðŸŽ¯ Tujuan Scatter Plot:

* Melihat **korelasi** antar dua variabel (positif, negatif, atau tidak ada).
* Mengidentifikasi **pola**, **kelompok (cluster)**, atau **outlier** dalam data.

---

### ðŸ“ˆ Contoh Visual:

Misalnya kamu punya data tinggi dan berat badan:

| Tinggi (cm) | Berat (kg) |
| ----------- | ---------- |
| 160         | 55         |
| 165         | 60         |
| 170         | 65         |
| 175         | 72         |

Scatter plot-nya akan menampilkan 4 titik, di mana:

* Sumbu X = tinggi
* Sumbu Y = berat
  Dan titik-titik ini akan cenderung membentuk garis naik â†’ menunjukkan **korelasi positif**.

---

### âœ… Contoh Kode di Python:

```python
import matplotlib.pyplot as plt

tinggi = [160, 165, 170, 175]
berat = [55, 60, 65, 72]

plt.scatter(tinggi, berat)
plt.title("Tinggi vs Berat")
plt.xlabel("Tinggi (cm)")
plt.ylabel("Berat (kg)")
plt.show()
```

---

### ðŸ§  Kapan Gunakan Scatter Plot?

* Saat ingin melihat hubungan antar dua variabel numerik.
* Saat menganalisis regresi atau korelasi.
* Saat mencari outlier yang menyimpang dari pola umum.

---

### ðŸ“Š Perbedaan dengan Line Plot dan Bar Chart:

| Chart         | Data yang Cocok                | Contoh Penggunaan         |
| ------------- | ------------------------------ | ------------------------- |
| **Scatter**   | Dua variabel numerik           | Hubungan tinggi vs berat  |
| **Line Plot** | Data deret waktu (time-series) | Perkembangan suhu harian  |
| **Bar Chart** | Kategori vs angka              | Jumlah pelanggan per hari |

---


````python
import matplotlib.pyplot as plt

tinggi = [160, 165, 170, 175]
berat = [55, 60, 65, 72]

plt.scatter(tinggi, berat)
plt.title("Tinggi vs Berat")
plt.xlabel("Tinggi (cm)")
plt.ylabel("Berat (kg)")
plt.show()
````

````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.scatter(x, y)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.show()
````

Grafik ini bisa lebih bermakna jika kita bisa menambahkan warna dan juga mengubah ukuran titik-titiknya. Kita dapat melakukan ini dengan menggunakan parameter c dan s masing-masing dari fungsi scatter.

Kustomisasi yang tersedia untuk scatter adalah â€“

- s: ukuran penanda (bisa skalar atau array dengan ukuran sama dengan ukuran x atau y)
- c: warna urutan warna untuk spidol
- marker: gaya penanda
- linewidths: lebar batas penanda
- edgecolor: warna batas penanda
- alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)


Kita juga dapat menampilkan bilah warna menggunakan metode colorbar()

````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
x = data['day']
y = data['total_bill']

# plotting the data
plt.scatter(x, y, c=data['size'],  s=data['total_bill'], marker='D', alpha=0.5)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Total Bill')

# Adding label on the x-axis
plt.xlabel('Day')

plt.colorbar()

plt.show()
````

# Pie Chart

### ðŸ¥§ Apa Itu **Pie Chart**?

**Pie chart** (diagram lingkaran) adalah grafik berbentuk **lingkaran** yang menunjukkan **proporsi atau persentase** dari total keseluruhan.
Setiap potongan (slice) mewakili bagian dari keseluruhan data.

---

### ðŸŽ¯ Tujuan Pie Chart:

* Menampilkan **perbandingan antar kategori** dalam satu kelompok.
* Menunjukkan **berapa besar kontribusi masing-masing kategori** terhadap total.

---

### ðŸ“Š Contoh Sederhana:

Misal kamu punya data jenis kelamin pasien di rumah sakit:

| Jenis Kelamin | Jumlah |
| ------------- | ------ |
| Laki-laki     | 60     |
| Perempuan     | 40     |

Pie chart-nya akan menampilkan:

* 60% lingkaran = laki-laki
* 40% lingkaran = perempuan

---

### âœ… Contoh Kode di Python:

```python
import matplotlib.pyplot as plt

labels = ['Laki-laki', 'Perempuan']
data = [60, 40]

plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Distribusi Jenis Kelamin Pasien")
plt.axis('equal')  # Biar pie-nya bulat sempurna
plt.show()
```

---

### ðŸ“Œ Kapan Gunakan Pie Chart?

* Saat kamu ingin menekankan **perbandingan proporsi**.
* Saat jumlah kategori **sedikit** (idealnya â‰¤ 5).
* Contoh: distribusi pengguna berdasarkan perangkat, gender, platform, dsb.

---

### âš ï¸ Catatan Penting:

* Pie chart **kurang cocok** jika ada **banyak kategori** atau perbedaan nilainya **kecil**.
* Kadang **bar chart** lebih efektif untuk perbandingan yang presisi.

---

### ðŸ“Ž Ringkasan:

| Chart         | Cocok Untuk                           | Contoh                              |
| ------------- | ------------------------------------- | ----------------------------------- |
| **Pie Chart** | Persentase dari satu total            | Distribusi gender, browser, dsb     |
| **Bar Chart** | Membandingkan kategori secara presisi | Jumlah pengunjung per hari/kategori |

---


````python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('tips.csv')

# initializing the data
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

# plotting the data
plt.pie(data, labels=cars)

# Adding title to the plot
plt.title("Car data")

plt.show()
````

Kustomisasi yang tersedia untuk pie chart adalah â€“

- explode: Memindahkan irisan plot
- autopct: Beri label baji dengan nilai numeriknya.
- color: Atribut digunakan untuk memberikan warna pada irisan.
- shadow: Digunakan untuk membuat bayangan baji.

````python
import matplotlib.pyplot as plt
import pandas as pd


# initializing the data
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 13, 35, 15, 12]

explode = [0.1, 0.5, 0, 0, 0]

colors = ( "orange", "cyan", "yellow",
		"grey", "green",)

# plotting the data
plt.pie(data, labels=cars, explode=explode, autopct='%.3f%%',
		colors=colors, shadow=True)

plt.show()
````
