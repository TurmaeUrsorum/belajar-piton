# Types of Data Visualization

**# Types of Data Visualization** berarti "**Jenis-jenis Visualisasi Data**". Ini adalah kategori atau cara berbeda dalam menampilkan data secara visual agar mudah dipahami.

Visualisasi data membantu kita melihat pola, tren, dan hubungan dalam data yang mungkin sulit terlihat hanya dalam bentuk tabel atau teks.

Berikut adalah beberapa **jenis visualisasi data yang umum**:

---

### 1. **Chart (Diagram/Grafik)**

* **Bar Chart (Diagram Batang)**: untuk membandingkan antar kategori.
* **Line Chart (Diagram Garis)**: menunjukkan tren dari waktu ke waktu.
* **Pie Chart (Diagram Lingkaran)**: menunjukkan proporsi/komposisi.
* **Area Chart**: mirip diagram garis, tapi dengan area berwarna di bawah garis.

---

### 2. **Plot**

* **Scatter Plot**: untuk menunjukkan hubungan antara dua variabel.
* **Bubble Chart**: seperti scatter plot, tapi ukuran titik menunjukkan variabel ketiga.

---

### 3. **Diagram Khusus**

* **Histogram**: untuk distribusi data (sering digunakan untuk statistik).
* **Box Plot (Diagram Kotak)**: menunjukkan median, kuartil, dan outlier.
* **Heatmap**: visualisasi data dalam bentuk warna (biasa untuk korelasi atau intensitas).

---

### 4. **Geographical Maps (Peta)**

* **Choropleth Map**: peta yang menunjukkan data berdasarkan warna wilayah (misalnya, kepadatan penduduk).
* **Dot Map**: menunjukkan lokasi titik-titik data.

---

### 5. **Hierarki & Jaringan**

* **Tree Map**: untuk menampilkan struktur hierarki dengan kotak berukuran proporsional.
* **Network Graph**: untuk relasi antar entitas (seperti jejaring sosial).

---

### 6. **Infografis dan Dashboard**

* Gabungan dari berbagai jenis visualisasi untuk menyampaikan informasi secara ringkas dan menarik.

# Line Plot

**Line plot** atau **line chart** adalah jenis visualisasi data yang menggunakan **garis lurus yang menghubungkan titik-titik data** untuk menunjukkan **perubahan nilai dari waktu ke waktu** atau berdasarkan urutan tertentu.

---

### 🔹 **Ciri-ciri Line Plot:**

* Sumbu **X** biasanya menunjukkan **waktu** atau urutan data (hari, bulan, tahun, dsb.).
* Sumbu **Y** menunjukkan **nilai atau ukuran** (seperti suhu, penjualan, jumlah pengguna, dll).
* Titik-titik data dihubungkan oleh **garis** untuk memperlihatkan tren atau pola.

---

### 📊 **Contoh Sederhana:**

Misalnya kamu ingin menunjukkan jumlah penjualan selama 5 hari:

| Hari   | Penjualan |
| ------ | --------- |
| Senin  | 10        |
| Selasa | 15        |
| Rabu   | 8         |
| Kamis  | 20        |
| Jumat  | 18        |

Line plot-nya akan menampilkan lima titik (satu untuk tiap hari), dan garis yang menghubungkan mereka akan memperlihatkan bagaimana penjualan naik-turun dari hari ke hari.

---

### ✅ **Kapan Pakai Line Plot?**

Gunakan **line plot** saat kamu ingin:

* Menunjukkan **tren dari waktu ke waktu**.
* Membandingkan **beberapa seri data waktu** (misalnya, penjualan 2 produk).
* Melihat **perubahan secara kontinu**.

---

### 🔧 Tools untuk Membuat Line Plot:

* **Excel**
* **Google Sheets**
* **Python** (pakai `matplotlib` atau `seaborn`)
* **Tableau, Power BI**, dan software visualisasi lainnya

---


````python
import pandas as pd 
import matplotlib.pyplot as plt 

# reading the database
data = pd.read_csv("tips.csv")

plt.figure(figsize=(10,5))
# Scatter plot with day against tip
# plt.plot(data['tip'], marker = 'o', linestyle='-.')
plt.plot(data["day"], data["tip"], marker = "o", linestyle="-.")
# plt.plot(data['size'])

# Adding Title to the Plot
plt.title("Line Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
````
