# Seaborn 

### 🎨 Apa Itu **Seaborn**?

**Seaborn** adalah **library visualisasi data di Python** yang dibangun di atas **Matplotlib**, dan dirancang untuk membuat grafik yang **lebih cantik, informatif, dan mudah digunakan** — terutama untuk **analisis data statistik**.

---

### 🔧 Kelebihan Seaborn:

✅ **Visual yang elegan secara default**
✅ **Integrasi langsung dengan `pandas DataFrame`**
✅ **Plot statistik kompleks lebih mudah dibuat**
✅ **Bisa membuat grafik seperti:**

| Jenis Plot   | Fungsi Seaborn      |
| ------------ | ------------------- |
| Bar Plot     | `sns.barplot()`     |
| Line Plot    | `sns.lineplot()`    |
| Scatter Plot | `sns.scatterplot()` |
| Histogram    | `sns.histplot()`    |
| Box Plot     | `sns.boxplot()`     |
| Heatmap      | `sns.heatmap()`     |
| Pairplot     | `sns.pairplot()`    |

---

### ✅ Contoh Penggunaan:

Misal kamu punya dataset tips (tips.csv) dari restoran:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("tips.csv")

# Tampilkan scatter plot antara total_bill dan tip
sns.scatterplot(data=data, x="total_bill", y="tip")

plt.title("Hubungan Total Tagihan dan Tip")
plt.show()
```

---

### 📌 Seaborn vs Matplotlib vs Pandas

| Library        | Fungsi Utama                     | Karakteristik                       |
| -------------- | -------------------------------- | ----------------------------------- |
| **Matplotlib** | Dasar semua visualisasi          | Sangat fleksibel tapi verbose       |
| **Pandas**     | Visualisasi cepat dari DataFrame | Cocok untuk eksplorasi sederhana    |
| **Seaborn**    | Statistik + visualisasi elegan   | Gaya default bagus, sintaks ringkas |

---

### 🧠 Kapan Gunakan Seaborn?

Gunakan Seaborn jika:

* Kamu bekerja dengan **DataFrame pandas**
* Ingin visualisasi **statistik** yang informatif
* Ingin hasil grafik yang **langsung rapi dan estetis**

## Mengubah estetika figur

set_style() digunakan untuk mengatur estetika plot. Artinya mempengaruhi hal-hal seperti warna sumbu, apakah grid aktif atau tidak, atau elemen estetika lainnya. Ada lima tema yang tersedia di Seaborn.

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# changing the theme to dark
sns.set_style("dark")
plt.show()
````

## Penghapusan spine

spines adalah garis yang mencatat batas data dan menghubungkan tanda centang sumbu. Itu dapat dihapus menggunakan metode despine().

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Removing the spines
sns.despine()
plt.show()
````

# Mengubah Ukuran Figur
Ukuran gambar dapat diubah menggunakan metode figure() dari Matplotlib. figure() metode membuat gambar baru dengan ukuran tertentu yang diteruskan dalam parameter figsize.

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# changing the figure size
plt.figure(figsize = (2, 4))

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Removing the spines
sns.despine()

plt.show()
````

## Mengatur Palet Warna default
set_palette() metode digunakan untuk mengatur palet warna default untuk semua plot. Argumen untuk color_palette() dan set_palette() sama. set_palette() mengubah parameter matplotlib default.

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

def plot():
	sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# setting the default color palette
sns.set_palette('vlag')
plt.subplot(211)

# plotting with the color palette
# as vlag
plot()

# setting another default color palette
sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
````

# Multipleplots Seaborn

Di **Seaborn**, `# Multipleplots` mengacu pada **pembuatan beberapa plot/grafik dalam satu tampilan**. Ini berguna saat kamu ingin membandingkan data berdasarkan kategori tertentu, atau ingin menampilkan beberapa aspek data secara berdampingan.

Seaborn menyediakan beberapa **fungsi khusus** dan **cara layout** untuk membuat multiple plots dengan mudah.

---

## 🔧 Cara Membuat Multiple Plots di Seaborn

### 1. **Menggunakan `FacetGrid`**

Digunakan untuk membuat grid (baris x kolom) berdasarkan kategori.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Buat grid berdasarkan kolom 'sex'
g = sns.FacetGrid(tips, col="sex")
g.map(sns.histplot, "total_bill")
plt.show()
```

> Ini akan membuat 2 histogram: satu untuk pria, satu untuk wanita.

---

### 2. **Menggunakan `catplot()`**

Fungsi all-in-one yang juga pakai `FacetGrid` di balik layar.

```python
sns.catplot(x="day", y="total_bill", col="sex", kind="box", data=tips)
plt.show()
```

> Ini menampilkan boxplot `total_bill` berdasarkan `day`, dipisah per `sex`.

---

### 3. **`pairplot()`**

Membuat plot matriks untuk melihat hubungan antar variabel numerik.

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

> Cocok untuk eksplorasi data: scatterplot antar semua kombinasi variabel.

---

### 4. **`subplot()` Manual (Matplotlib + Seaborn)**

Kalau ingin kendali penuh, bisa pakai `plt.subplot()` lalu pasang seaborn plot di masing-masing:

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(tips[tips["sex"] == "Male"]["total_bill"], ax=axes[0])
sns.histplot(tips[tips["sex"] == "Female"]["total_bill"], ax=axes[1])
axes[0].set_title("Laki-laki")
axes[1].set_title("Perempuan")
plt.show()
```

---

`# Multipleplots` di Seaborn sangat berguna untuk:

## Metode 1: Menggunakan metode FacetGrid()



- Kelas FacetGrid membantu dalam memvisualisasikan distribusi satu variabel serta hubungan antara beberapa variabel secara terpisah dalam himpunan bagian dari kumpulan data Anda menggunakan beberapa panel.
- Sebuah FacetGrid dapat digambar hingga tiga dimensi ? baris, col, dan hue. Dua yang pertama memiliki korespondensi yang jelas dengan susunan sumbu yang dihasilkan; pikirkan variabel rona sebagai dimensi ketiga di sepanjang sumbu kedalaman, di mana tingkat yang berbeda diplot dengan warna yang berbeda.
- Objek FacetGrid mengambil dataframe sebagai input dan nama-nama variabel yang akan membentuk dimensi baris, kolom, atau hue dari grid. Variabel harus kategoris dan data pada setiap tingkat variabel akan digunakan untuk segi di sepanjang sumbu itu.

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")

plt.show()
````

## Metode 2: Menggunakan metode PairGrid()



- subplot grid untuk memplot hubungan berpasangan dalam kumpulan data.
- Kelas ini memetakan setiap variabel dalam kumpulan data ke kolom dan baris dalam grid beberapa sumbu. Fungsi plotting tingkat sumbu yang berbeda dapat digunakan untuk menggambar plot bivariat di segitiga atas dan bawah, dan distribusi marjinal dari setiap variabel dapat ditunjukkan pada diagonal.
- Ini juga dapat mewakili tingkat konvensionalisasi tambahan dengan parameter hue, yang memplot subset data yang berbeda dalam warna yang berbeda. Ini menggunakan warna untuk menyelesaikan elemen pada dimensi ketiga, tetapi hanya menggambar subset di atas satu sama lain dan tidak akan menyesuaikan parameter hue untuk visualisasi spesifik seperti yang dilakukan fungsi tingkat sumbu yang menerima hue.


````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("flights")

plot = sns.PairGrid(data)
plot.map(plt.plot)

plt.show()
````

**Categorical plot** (atau *plot kategorikal*) adalah jenis visualisasi data yang digunakan untuk menampilkan hubungan antara **data kategorikal** (data yang terdiri dari kategori atau label seperti "laki-laki/perempuan", "ya/tidak", "warna", dll.) dengan variabel lainnya, biasanya numerik.

### Tujuan

Membantu memahami distribusi, frekuensi, perbandingan, atau hubungan antar kategori.

---

### Jenis-Jenis Categorical Plot (paling umum digunakan):

1. **Bar Plot**

   * Menampilkan nilai agregat (seperti rata-rata, jumlah, dll.) dari data numerik berdasarkan kategori.
   * Contoh: rata-rata gaji berdasarkan jenis kelamin.

2. **Count Plot**

   * Menunjukkan jumlah kemunculan (frekuensi) dari tiap kategori.
   * Cocok jika kamu ingin tahu berapa banyak data dalam setiap kategori.

3. **Box Plot**

   * Menunjukkan distribusi data numerik berdasarkan kategori.
   * Berguna untuk melihat median, kuartil, outlier, dll.

4. **Violin Plot**

   * Gabungan antara box plot dan density plot.
   * Menunjukkan distribusi data dan kepadatannya berdasarkan kategori.

5. **Strip Plot**

   * Menampilkan semua titik data, cocok untuk dataset kecil.
   * Sering digunakan bersama swarm plot.

6. **Swarm Plot**

   * Versi strip plot tapi tidak tumpang tindih; memberikan gambaran distribusi data yang rapi.

---

### Kapan Digunakan?

* Saat ingin menganalisis hubungan antara **kategori** dan **nilai numerik**.

Kalau kamu pakai tools seperti Excel, pandas, Seaborn, atau Tableau, categorical plot sangat umum dipakai dalam dashboard atau laporan analitik.


## Categorical Plot


Plot Kategoris digunakan di mana kita harus memvisualisasikan hubungan antara dua nilai numerik. Pendekatan yang lebih khusus dapat digunakan jika salah satu variabel utama adalah kategorikal yang berarti variabel-variabel tersebut mengambil sejumlah nilai yang mungkin tetap dan terbatas.

### boxplot

Ini menunjukkan distribusi data kuantitatif yang mewakili perbandingan antar variabel. boxplot menunjukkan kuartil dari dataset sementara whiskers meluas untuk menunjukkan sisa distribusi yaitu titik-titik yang menunjukkan adanya outlier. Itu dibuat menggunakan metode boxplot().

````python
import seaborn as sns 
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.boxplot(x="species", y="sepal_length", data=data)
plt.show()
````

### violin plot
Ini mirip dengan boxplot kecuali bahwa ia memberikan visualisasi yang lebih tinggi dan lebih maju dan menggunakan estimasi kepadatan kernel untuk memberikan deskripsi yang lebih baik tentang distribusi data. Itu dibuat menggunakan metode violinplot().

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
````

### Strip plot


Ini pada dasarnya membuat plot pencar berdasarkan kategori. Itu dibuat menggunakan metode stripplot().

````python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.stripplot(x='species', y='sepal_width', data=data)
plt.show()
````

**Plot distribution** (atau *plot distribusi*) adalah visualisasi yang digunakan untuk menunjukkan **penyebaran (distribusi)** nilai-nilai dalam suatu dataset, khususnya untuk data numerik.

---

### Tujuan

Membantu kamu memahami:

* Seberapa sering nilai tertentu muncul
* Pola distribusi (apakah normal, skewed, multimodal, dll.)
* Apakah ada outlier (nilai yang menyimpang jauh)
* Bentuk data: simetris, miring ke kiri/kanan, rata, tajam, dll.

---

### Jenis-Jenis Plot Distribution (umum digunakan):

1. **Histogram**

   * Membagi nilai-nilai menjadi *bin* (kelompok), lalu menghitung frekuensinya.
   * Contoh: berapa banyak orang yang memiliki tinggi antara 160–170 cm.

2. **KDE Plot (Kernel Density Estimate)**

   * Kurva halus yang memperkirakan distribusi probabilitas dari data.
   * Alternatif yang lebih halus dari histogram.

3. **Box Plot**

   * Menampilkan ringkasan lima angka (minimum, kuartil 1, median, kuartil 3, maksimum) dan outlier.
   * Baik untuk membandingkan distribusi antar grup.

4. **Violin Plot**

   * Kombinasi KDE plot + box plot.
   * Menunjukkan bentuk distribusi dan statistik ringkasan sekaligus.

5. **ECDF Plot (Empirical Cumulative Distribution Function)**

   * Menunjukkan proporsi kumulatif data sampai nilai tertentu.

---

### Contoh Kode (Python dengan Seaborn/Matplotlib)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Histogram
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Distribusi Total Bill")
plt.show()
```

---

### Kapan Digunakan?

* Saat melakukan **Exploratory Data Analysis (EDA)**
* Untuk mengecek apakah data mengikuti distribusi normal
* Saat ingin tahu bagaimana sebaran skor, harga, jumlah, dll. dalam suatu kolom

---

### Perbedaan dengan Categorical Plot:

* **Plot distribution** → fokus ke **penyebaran nilai numerik**
* **Categorical plot** → fokus ke **perbandingan antar kategori**

## Regression Plot

Plot regresi terutama dimaksudkan untuk menambahkan panduan visual yang membantu untuk menekankan pola dalam kumpulan data selama analisis data eksplorasi. Plot regresi seperti namanya menciptakan garis regresi antara dua parameter dan membantu memvisualisasikan hubungan liniernya.

implot

Metode lmplot() dapat dipahami sebagai fungsi yang pada dasarnya membuat plot model linier. Ini menciptakan plot pencar dengan kecocokan linier di atasnya.

````python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

sns.lmplot(x='total_bill', y='tip', data=data)
plt.show()
````
